# Importação das bibliotecas

```python
import pandas as pd
from google import colab as cl
import plotly.express as px
```

```python
file_upload = cl.files.upload()
df = pd.read_csv('train.csv')
```

# Limpeza dos Dados

```python
df1 = df.copy()

# convertendo a coluna Age de texto para Númmero
selected_rows = (df1['Delivery_person_Age'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()

df1['Delivery_person_Age'] = df1['Delivery_person_Age'].astype(int)
df1.shape

# convertendo a coluna Ratings de texto para número decimal
df1['Delivery_person_Ratings'] = df1['Delivery_person_Ratings'].astype(float)

# convertendo a coluna Order_Date de texto para data
df1['Order_Date'] = pd.to_datetime(df1['Order_Date'], format='%d-%m-%Y')

# convertendo multiplee_deliveries de texto para número int
selected_rows = (df1['multiple_deliveries'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()
df1['multiple_deliveries'] = df1['multiple_deliveries'].astype(int)

# removendo "NaNs" das colunas Weatherconditions, City, Festival, Road_traffic_density, Delivery_person_Age
selected_rows = (df1['Weatherconditions'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()
selected_rows = (df1['Weatherconditions'] != 'conditions NaN')

selected_rows = (df1['City'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()

selected_rows = (df1['Festival'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()

selected_rows = (df1['Road_traffic_density'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()

# removing space at the end of strinf from cols Type_of_vehicle, Festival
df1['Type_of_vehicle'] = df1['Type_of_vehicle'].str.rstrip()
df1['Festival'] = df1['Festival'].str.rstrip()
df1['Road_traffic_density'] = df1['Road_traffic_density'].str.rstrip()
df1['Type_of_order'] = df1['Type_of_order'].str.rstrip()
```

# Exercícios

## 1. A menor e maior idade dos entregadores

```python
oldest_deliverer = df1.loc[:, 'Delivery_person_Age'].max()
youngst_deliverer = df1.loc[:, 'Delivery_person_Age'].min()

print(f'A maior idade dos entregadores é: {oldest_deliverer}, enquanto a menor idade é: {youngst_deliverer}')
```

## 2. A pior e a melhor condição de veículos

```python
lowest_vehicle_condition = df1.loc[:, 'Vehicle_condition'].min()

print(f'A pior condição de veículo é: {lowest_vehicle_condition}')
```

## 3. A avaliação média por entregador

```python
df_aux = df1.loc[:, ['Delivery_person_ID','Delivery_person_Ratings']].groupby('Delivery_person_ID').mean().reset_index()

df_aux
```

## 4. A avaliação média e o desvio padrão por tipo de tráfego.

```python
df_aux = df1.loc[:, ['Road_traffic_density', 'Delivery_person_Ratings']]

tabelized = df_aux.groupby('Road_traffic_density').agg({'Delivery_person_Ratings' : ['mean', 'std']})

tabelized.columns = ['delivery_mean', 'delivery_std']

tabelized.reset_index()
```

## 5. A avaliação médio e o desvio padrão por condições climáticas

```python
df_aux = df1.loc[:, ['Weatherconditions', 'Delivery_person_Ratings']]

tabelized = df_aux.groupby('Weatherconditions').agg({'Delivery_person_Ratings' : ['mean','std']})

tabelized.columns = ['delivery_mean', 'delivery_std']

tabelized.reset_index()
```

## 6. Os 10 entregadores mais rápidos por cidade

```python
df_aux = df1.loc[:, ['City', 'Delivery_person_ID', 'Time_taken(min)']]

df_aux = df_aux.groupby(['City', 'Delivery_person_ID']).min().sort_values(['City','Time_taken(min)']).reset_index()

cities = df1['City'].unique()

df_result = pd.DataFrame()

for c in cities:
  df_aux_city = df_aux.loc[df_aux['City'] == c, :].head(10)
  df_result = pd.concat([df_result, df_aux_city])


df_result
```

## 7. Os 10 entregadores mais lentos por cidade

```python
df_aux = df1.loc[:, ['City', 'Delivery_person_ID', 'Time_taken(min)']]

df_aux = df_aux.groupby(['City', 'Delivery_person_ID']).min().sort_values(['City','Time_taken(min)']).reset_index()

cities = df1['City'].unique()

df_result = pd.DataFrame()

for c in cities:
  df_aux_city = df_aux.loc[df_aux['City'] == c, :].tail(10)
  df_result = pd.concat([df_result, df_aux_city])


df_result
```
