# Importação das Bibliotecas

```python
import pandas as pd
from google import colab as cl
import plotly.express as px
```

# Importação dos Dados

```python
file_upload = cl.files.upload()
df = pd.read_csv('train.csv')
```

# Limpeza dos Dados

```python
df1 = df.copy()

# convertendo a coluna Age de texto para Númmero
selected_rows = (df1['Delivery_person_Age'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()

df1['Delivery_person_Age'] = df1['Delivery_person_Age'].astype(int)
df1.shape

# convertendo a coluna Ratings de texto para número decimal
df1['Delivery_person_Ratings'] = df1['Delivery_person_Ratings'].astype(float)

# convertendo a coluna Order_Date de texto para data
df1['Order_Date'] = pd.to_datetime(df1['Order_Date'], format='%d-%m-%Y')

# convertendo multiplee_deliveries de texto para número int
selected_rows = (df1['multiple_deliveries'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()
df1['multiple_deliveries'] = df1['multiple_deliveries'].astype(int)

# removendo "NaNs" das colunas Weatherconditions, City, Festival, Road_traffic_density
selected_rows = (df1['Weatherconditions'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()
selected_rows = (df1['Weatherconditions'] != 'conditions NaN')

selected_rows = (df1['City'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()

selected_rows = (df1['Festival'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()

selected_rows = (df1['Road_traffic_density'] != 'NaN ')
df1 = df1.loc[selected_rows, :].copy()

# removing space at the end of strinf from cols Type_of_vehicle, Festival
df1['Type_of_vehicle'] = df1['Type_of_vehicle'].str.rstrip()
df1['Festival'] = df1['Festival'].str.rstrip()
df1['Road_traffic_density'] = df1['Road_traffic_density'].str.rstrip()
df1['Type_of_order'] = df1['Type_of_order'].str.rstrip()
```

# Exercícios

## 1. Desenhar um gráfico de pizza com a média de idade dos entregadores por cidade

```python
 df_aux = df1.loc[:, ['Delivery_person_Age', 'City']].groupby('City').mean().reset_index()

px.pie(df_aux, values='Delivery_person_Age', names='City')
```

## 2. Desenhe um gráfico de linha, mostrando o número total de entregas diárias feitas por densidade de tráfego igual a "Low" e "Jam"

```python
cols = ['Road_traffic_density','ID', 'Order_Date']

filtered_df = df1[(df1['Road_traffic_density'] == 'Low') | (df1['Road_traffic_density'] == 'Jam')]

df_aux = filtered_df.loc[:, cols].copy()

grouped = df_aux.groupby(['Road_traffic_density', 'Order_Date'])['ID'].count().reset_index()

grouped

#px.line(grouped, x='Order_Date', y='ID')
```

## 3. Desenha um gráfico de barras, mostrando as avaliações médias das entregas por semana.

```python
df1['week_of_year'] = df1['Order_Date'].dt.strftime('%U')

df_aux = df1.loc[:, ['ID', 'week_of_year']].groupby('week_of_year')#.count().reset_index()

df_aux
#px.bar(df_aux, x='week_of_year', y='ID')
```

## 4. Desenhe um gráfico de pizza, mostrando a média de avaliações das entregas feitas por condições climáticas

```python

df_aux = df1.loc[:, ['Weatherconditions', 'Delivery_person_Ratings']]

grouped = df_aux.groupby('Weatherconditions')['Delivery_person_Ratings'].median().reset_index()

px.pie(grouped, values='Delivery_person_Ratings', names='Weatherconditions')
```
